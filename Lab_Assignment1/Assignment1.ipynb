```python
# step 1
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# step 2
# Load data.xlsx file, which contains the observations (with noise), variance of observations and true observations
# hint: use pandas library to parse excel file and use numpy library to convert it to array
# (1 mark)

url_data = "https://raw.githubusercontent.com/NazilaAbedini/ENGG680_2025_Fall/main/Lab_Assignment1/data.xlsx"

df_data = pd.read_excel(url_data)
arr_data = df_data.to_numpy()

#store the noisy observations, variance of observations and true observations into numpy arrays seperately
# (1 mark)
l = arr_data[:, 0]       # noisy observations (1st column)
l_var = arr_data[:, 1]   # variance of observations (2nd column)
l_true = arr_data[:, 2]  # true observations (3rd column)
# Display to verify
print("Data loaded successfully:")
print(df_data.head())
```

```python
# step 3
# define the number of observation n
# (1 mark)
n = len(l)

# define the observation weight matrix P for LSE as identity matrix
# identity weight matrix simply means the observations will be unweighted in the following LSE
# (1 mark)
P = np.eye(n)

# define the range of function input(i.e., constants a), from start to end; and the numerical array of a
start = -1* np.pi
end = np.pi
a = np.linspace(start, end, n)
# Display to verify
print("Number of observations (n):", n)
print("Shape of weight matrix (P):", P.shape)
print("Range of a:", a[:5], "...", a[-5:])
```

```python
# step 4
# use the matplotlib to plot the true and noisy observations together in one single figure
# use scatter for noisy observation plotting
# for the true observation, label it as "f(a)"" and set color as 'k'
# for the noisy observation scatter plot, label it as "f(a) + e", set color as 'tab:blue', set marker as 'o' and set alpha as 0.3
# include proper axis labels, legend and title
# scale the range of x-axis to be -1 to 1
# keep this format for the rest of the plots in this lab assignment
# (2 marks)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(14, 5))

# Plot the true observation line (f(a))
ax.plot(a, l_true, label='f(a)', color='k', linewidth=2)

# Plot the noisy observations as scatter points (f(a) + e)
ax.scatter(a, l, label='f(a) + e', color='tab:blue', marker='o', alpha=0.3)

# Add labels, title, legend, and grid
ax.set_xlabel('a', fontsize=12)
ax.set_ylabel('â„“', fontsize=12)
ax.set_title('True Function vs Noisy Observations', fontsize=14)
ax.legend()

# Scale x-axis range to -1 to 1
ax.set_xlim(-1, 1)

# Add grid and layout formatting

ax.grid(True, which='both',ls='-',color='0.65')
fig.tight_layout()

# Display the plot
plt.show()
```

```python
# step 5
# define a curve fitting function named fit_polynomial_curve
# the function should implement the linear least square estimation
# the function should include all important intermediate steps in LSE
# the input and output of the funciton are described in the comment below in red.
# please keep code clean and well-commented
# (7 marks)

def fit_polynomial_curve(a: np.ndarray, l: np.ndarray, P: np.ndarray, deg: int) -> tuple:
    '''
    Linear weighted least squares fit of a polynomial to noisy observations.

    Args:
        a   : (n,) array â€” function inputs (independent variable)
        l   : (n,) array â€” noisy observations (dependent variable)
        P   : (n,n) array â€” observation weight matrix (use I for unweighted)
        deg : int        â€” degree of the polynomial to fit

    Returns:
        x     : (deg+1,) array â€” estimated polynomial coefficients (x0 + x1 a + ... + x_deg a^deg)
        cov_x : (deg+1,deg+1) array â€” covariance matrix of the estimated coefficients
        apv   : float â€” a posteriori variance factor (goodness of fit)
        v     : (n,) array â€” residuals (A x - l)
    '''
    # ---- 1) Build the design (Vandermonde) matrix A: columns a^0, a^1, ..., a^deg
    # Shape: (n, deg+1)
    A = np.vander(a, N=deg+1, increasing=True)

    # ---- 2) Form normal equations: N = A^T P A, u = A^T P l
    AtP = A.T @ P
    N = AtP @ A
    u = AtP @ l

    # ---- 3) Solve for coefficients x (use solve if well-conditioned; fall back to pinv otherwise)
    try:
        x = np.linalg.solve(N, u)
    except np.linalg.LinAlgError:
        x = np.linalg.pinv(N) @ u  # robust fallback

    # ---- 4) Compute residuals v = A x - l
    v = A @ x - l

    # ---- 5) Compute a posteriori variance factor (goodness of fit)
    n_obs = A.shape[0]
    n_unk = A.shape[1]
    dof = max(n_obs - n_unk, 1)  # avoid division by zero
    apv = float(v.T @ P @ v) / dof

    # ---- 6) Covariance of the estimated coefficients
    try:
        N_inv = np.linalg.inv(N)
    except np.linalg.LinAlgError:
        N_inv = np.linalg.pinv(N)
    cov_x = apv * N_inv

    return (x, cov_x, apv, v)
```

```python
# step 6
# call the fit_polynomial_curve function and fetch the results
# use polynomial degrees of: 2, 4, 6
# (1 marks)
# Degree 2
x2, cov_x2, apv2, v2 = fit_polynomial_curve(a, l, P, deg=2)

# Degree 4
x4, cov_x4, apv4, v4 = fit_polynomial_curve(a, l, P, deg=4)

# Degree 6
x6, cov_x6, apv6, v6 = fit_polynomial_curve(a, l, P, deg=6)

# Display results for quick verification
print("Degree 2 coefficients:", x2)
print("Degree 4 coefficients:", x4)
print("Degree 6 coefficients:", x6)
```

```python
# step 7
# use the estimated coefficients x and constants array a, construct polynomial functions for each degree
# hint: use polyval function from numpy library (check input argument carefully)
# (1 marks)
# import numpy as np
quadratic_uw = np.polyval(x2[::-1], a)   # degree 2 fit
quartic_uw   = np.polyval(x4[::-1], a)   # degree 4 fit
hexic_uw     = np.polyval(x6[::-1], a)   # degree 6 fit
```

```python
# step 8
# use the matplotlib to plot each polynomial function, noisy observations l, true observations l_true, together in one single figure
# for the polynomials of line, cubic and quintic, use color of orange, green and red respectively.
# (2 marks)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(14, 5))
# Plot the true function
ax.plot(a, l_true, label='True f(a)', color='k', linewidth=2)

# Plot noisy observations
ax.scatter(a, l, label='Noisy Observations f(a) + e', color='tab:blue', marker='o', alpha=0.3)

# Plot polynomial fits
ax.plot(a, quadratic_uw, color='orange', linewidth=2, label='Quadratic Fit (deg=2)')
ax.plot(a, quartic_uw, color='green', linewidth=2, label='Quartic Fit (deg=4)')
ax.plot(a, hexic_uw, color='red', linewidth=2, label='Hexic Fit (deg=6)')

# Labels, title, and legend
ax.set_xlabel('a', fontsize=12)
ax.set_ylabel('â„“', fontsize=12)
ax.set_title('Polynomial Fits of Various Degrees vs True and Noisy Observations', fontsize=14)
ax.legend(loc='upper left')

# Grid and layout

ax.grid(True, which='both',ls='-',color='0.65')

fig.tight_layout()
```
